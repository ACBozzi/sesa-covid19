```python
import pandas as pd
from hashlib import md5
from bulletin.commom.normalize import *

def normalize_do(do):
    try:
        do = str(do).replace("-","")
        return int(do)
    except:
        return None
```

```python
obitos_novos = pd.read_excel('novos_obitos.xlsx')
obitos_novos.columns = [ normalize_text(x) for x in obitos_novos.columns ]

casos = pd.read_excel('Casos confirmados PR.xlsx','Casos confirmados', usecols='B:M')
casos.columns = [ normalize_text(x) for x in casos.columns ]

obitos = pd.read_excel('Casos confirmados PR.xlsx','Obitos', usecols='B:I')
obitos.columns = [ normalize_text(x) for x in obitos.columns ]
```

```python
obitos_novos = obitos_novos.rename(columns={'nome_paciente':'nome','municipio_residencia':'mun_resid','municipio_atendimento':'mun_atend','data_diagnostico':'dt_diag','numero_do':'do'})

# obitos_novos['nome'] = obitos_novos['nome'].apply(normalize_text)
# obitos_novos['sexo'] = obitos_novos['sexo'].apply(normalize_text)
# obitos_novos['idade'] = obitos_novos['idade'].apply(lambda x: normalize_number(x,fill=0))
# obitos_novos['mun_resid'] = obitos_novos['mun_resid'].apply(normalize_municipios)
# obitos_novos['mun_atend'] = obitos_novos['mun_atend'].apply(normalize_municipios)
# obitos_novos['laboratorio'] = obitos_novos['laboratorio'].apply(normalize_text)
# obitos_novos['dt_diag'] = obitos_novos['dt_diag'].apply(normalize_date)
obitos_novos['do'] = obitos_novos['do'].apply(normalize_do)

obitos_novos['hash'] = obitos_novos.apply( lambda row: md5(str.encode( normalize_text(row['nome']) + 
                                                                      str(normalize_number(row['idade'],fill=0)) + 
                                                                      normalize_text(row['mun_resid']) )).hexdigest(), axis=1)
```

```python
casos = casos.rename(columns={'comunicacao':'dt_com','ibge_res_pr':'ibge'})

# casos['nome'] = casos['nome'].apply(normalize_text)
# casos['sexo'] = casos['sexo'].apply(normalize_text)
# casos['idade'] = casos['idade'].apply(lambda x: normalize_number(x,fill=0))
# casos['ibge'] = casos['ibge'].apply(normalize_igbe)
casos['mun_resid'] = casos['mun_resid'].apply(normalize_municipios)
# casos['mun_atend'] = casos['mun_atend'].apply(normalize_municipios)
# casos['laboratorio'] = casos['laboratorio'].apply(normalize_text)
# casos['dt_com'] = casos['dt_com'].apply(normalize_date)
# casos['dt_diag'] = casos['dt_diag'].apply(normalize_date)

casos = casos.loc[casos['mun_resid']!='excluir']

casos['hash'] = casos.apply( lambda row: md5(str.encode( normalize_text(row['nome']) + 
                                                                      str(normalize_number(row['idade'],fill=0)) + 
                                                                      normalize_text(row['mun_resid']) )).hexdigest(), axis=1)
```

```python
obitos = obitos.rename(columns={'ibge_res_pr':'ibge','municipio':'mun_resid','data_do_obito':'dt_obt'})

# obitos['nome'] = obitos['nome'].apply(normalize_text)
# obitos['sexo'] = obitos['sexo'].apply(normalize_text)
# obitos['idade'] = obitos['idade'].apply(normalize_number)
# obitos['ibge'] = obitos['ibge'].apply(normalize_igbe)
obitos['mun_resid'] = obitos['mun_resid'].apply(normalize_municipios)
# obitos['dt_obt'] = obitos['dt_obt'].apply(normalize_date)

obitos = obitos.loc[obitos['mun_resid']!='excluir']

obitos['hash'] = obitos.apply( lambda row: md5(str.encode( normalize_text(row['nome']) + 
                                                                      str(normalize_number(row['idade'],fill=0)) + 
                                                                      normalize_text(row['mun_resid']) )).hexdigest(), axis=1)
```
